```python
list1=[ ]
list2=[ ]
n=int(input("how many elements  :"))
for i in range(n):
    a=int(input("enter length:"))
    list1.append(a)
m=int(input("how many elements  :"))
for i in range(m):
    b=int(input("enter breadth:"))
    list2.append(b)
dict={ }
value=[ ]
key=[ ]
for i in range(n):
    area=list1[i]*list2[i]
    value.append(area)
    key.append(i+1)
    dict[key[i]]=value[i]
print("dictionary is:",dict)
print("maximum area=",max(value))
```
